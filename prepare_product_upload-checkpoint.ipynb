```python
import pandas as pd
import os
import re
import csv
```

```python
"""
Input: Input csv with Full name of the perfume and vendor as per website standard
Output: Output csv that follows product export format
"""
```

```python
home_dir = os.path.expanduser("~")
input_path = os.path.join(home_dir, "Downloads", "product_names.csv")
output_path = os.path.join(home_dir, "Downloads", "product_export_v0.csv")
print(input_path, output_path)
```

```python
output_columns = [
    "Handle", "Title", "Body (HTML)", "Vendor", "Product Category", "Type", "Tags", "Published",
    "Option1 Name", "Option1 Value", "Option1 Linked To", "Option2 Name", "Option2 Value", "Option2 Linked To",
    "Option2 Linked To", "Option3 Name", "Option3 Value", "Option3 Linked To", "Variant SKU", "Variant Grams",
    "Variant Inventory Tracker", "Variant Inventory Policy", "Variant Fulfillment Service", "Variant Price",
    "Variant Compare At Price", "Variant Requires Shipping", "Variant Taxable", "Variant Barcode", "Image Src",
    "Image Position", "Image Alt Text", "Gift Card", "SEO Title", "SEO Description","Google Shopping / Google Product Category",
    "Google Shopping / Gender", "Google Shopping / Age Group", "Google Shopping / MPN", "Google Shopping / Condition",
    "Google Shopping / Custom Product", "Google Shopping / Custom Label 0", "Google Shopping / Custom Label 1",
    "Google Shopping / Custom Label 2","Google Shopping / Custom Label 3", "Google Shopping / Custom Label 4",
    "Google: Custom Product (product.metafields.mm-google-shopping.custom_product)", "Product rating count (product.metafields.reviews.rating_count)",
    "Color (product.metafields.shopify.color-pattern)", "Occasion (product.metafields.shopify.occasion)", "Scent (product.metafields.shopify.scent)",
    "Season (product.metafields.shopify.season)", "Target gender (product.metafields.shopify.target-gender)",
    "Complementary products (product.metafields.shopify--discovery--product_recommendation.complementary_products)",
    "Related products (product.metafields.shopify--discovery--product_recommendation.related_products)",
    "Related products settings (product.metafields.shopify--discovery--product_recommendation.related_products_display)",
    "Search product boosts (product.metafields.shopify--discovery--product_search_boost.queries)", "Variant Image",
    "Variant Weight Unit", "Variant Tax Code", "Cost per item", "Status"
]
```

```python
"""
1. Main perfume row
2. Main decant row
3. Decant variant rows
"""
```

```python
decant_variants = (
    (True,'5ml Decant', "https://cdn.shopify.com/s/files/1/0741/4474/3680/files/Product_page_template_27_2838f26f-31ec-47c7-9975-d24e0bb80983.png?v=1754821855", 1),
    (False,'10ml Decant', "https://cdn.shopify.com/s/files/1/0741/4474/3680/files/Product_page_template_28_26823e29-5144-48ab-9513-43024d1f3be2.png?v=1754821854", 2),
    (False,'20ml Decant', "https://cdn.shopify.com/s/files/1/0741/4474/3680/files/Product_page_template_30_5a29f289-1fac-4d53-911c-999cacbc1cd0.png?v=1754821881", 3),
    (False,'30ml Decant', "https://cdn.shopify.com/s/files/1/0741/4474/3680/files/Product_page_template_29_0fd0a356-03fb-40f4-b5f5-630052ed5c9c.png?v=1754821882", 4)
)
PRODUCT_CATAGORY = "Health & Beauty > Personal Care > Cosmetics > Perfumes & Colognes"
```

```python
with open(input_path, 'r', encoding='utf-8', errors='ignore', newline='') as infile, open(output_path, 'w', newline='') as outfile:
    reader = csv.DictReader(infile)
    writer = csv.DictWriter(outfile, fieldnames=output_columns)
    writer.writeheader()
   
    for row in reader:
        print(row)
        try:
            main_title_complete = False
            main_decant_complete = False
            decant_title = re.sub(r'\s*\d+ml\s*', ' ', row['title'])
            decant_title = re.sub(r'\s+',' ', decant_title).strip() + " Decant"
            decant_idx = 0

            for i in range(len(decant_variants)+1):
                output_map = {}
                if not main_title_complete:
                    for col in output_columns:
                        if col == "Handle":
                            output_map[col] = "-".join(row['title'].lower().split())
                        elif col == 'Title':
                            output_map[col] = row['title']
                        elif col == 'Vendor':
                            output_map[col] = row['vendor']
                        elif col == 'Product Category':
                            output_map[col] = PRODUCT_CATAGORY
                        elif col == 'Published':
                            output_map[col] = 'FALSE'
                        elif col == 'Option1 Name':
                            output_map[col] = 'Title'
                        elif col == 'Option1 Value':
                            output_map[col] = 'Default Title'
                        elif col == 'Variant Grams':
                            output_map[col] = 0
                        elif col == 'Variant Inventory Tracker':
                            output_map[col] = 'shopify'
                        elif col == 'Variant Inventory Policy':
                            output_map[col] = 'deny'
                        elif col == 'Variant Fulfillment Service':
                            output_map[col] = 'manual'
                        elif col == 'Variant Requires Shipping':
                            output_map[col] = 'TRUE'
                        elif col == 'Variant Taxable':
                            output_map[col] = 'TRUE'
                        elif col == 'Gift Card':
                            output_map[col] = 'FALSE'
                        elif col == 'Target gender (product.metafields.shopify.target-gender)':
                            output_map[col] = 'male;female'
                        elif col == 'Variant Weight Unit':
                            output_map[col] = 'kg'
                        elif col == 'Status':
                            output_map[col] = 'draft'
                        else:
                            output_map[col] = ""
                    
                    writer.writerow(output_map)
                    main_title_complete = True
                else:
                    if not main_decant_complete:
                        for col in output_columns:
                            if col == "Handle":
                                output_map[col] = "-".join(decant_title.lower().split())
                            elif col == 'Title':
                                output_map[col] = decant_title
                            elif col == 'Vendor':
                                output_map[col] = row['vendor']
                            elif col == 'Product Category':
                                output_map[col] = PRODUCT_CATAGORY
                            elif col == "Type":
                                output_map[col] = 'Decant'
                            elif col == 'Published':
                                output_map[col] = 'FALSE'
                            elif col == 'Option1 Name':
                                output_map[col] = 'Quantity'
                            elif col == 'Option1 Value':
                                output_map[col] = decant_variants[decant_idx][1]
                            elif col == "Image Src":
                                output_map[col] = decant_variants[decant_idx][2]
                            elif col == "Image Position":
                                output_map[col] = decant_variants[decant_idx][3]
                            elif col == 'Variant Grams':
                                output_map[col] = 0
                            elif col == 'Variant Inventory Tracker':
                                output_map[col] = 'shopify'
                            elif col == 'Variant Inventory Policy':
                                output_map[col] = 'deny'
                            elif col == 'Variant Fulfillment Service':
                                output_map[col] = 'manual'
                            elif col == 'Variant Requires Shipping':
                                output_map[col] = 'TRUE'
                            elif col == 'Variant Taxable':
                                output_map[col] = 'TRUE'
                            elif col == 'Gift Card':
                                output_map[col] = 'FALSE'
                            elif col == 'Target gender (product.metafields.shopify.target-gender)':
                                output_map[col] = 'male;female'
                            elif col == 'Variant Weight Unit':
                                output_map[col] = 'kg'
                            elif col == 'Status':
                                output_map[col] = 'draft'
                            else:
                                output_map[col] = ""
                        
                        writer.writerow(output_map)
                        main_decant_complete = True
                        decant_idx += 1
                    else:
                        for col in output_columns:
                            if col == "Handle":
                                output_map[col] = "-".join(decant_title.lower().split())
                            elif col == 'Option1 Value':
                                output_map[col] = decant_variants[decant_idx][1]
                            elif col == "Image Src":
                                output_map[col] = decant_variants[decant_idx][2]
                            elif col == "Image Position":
                                output_map[col] = decant_variants[decant_idx][3]
                            elif col == 'Variant Grams':
                                output_map[col] = 0
                            elif col == 'Variant Inventory Tracker':
                                output_map[col] = 'shopify'
                            elif col == 'Variant Inventory Policy':
                                output_map[col] = 'deny'
                            elif col == 'Variant Fulfillment Service':
                                output_map[col] = 'manual'
                            elif col == 'Variant Requires Shipping':
                                output_map[col] = 'TRUE'
                            elif col == 'Variant Taxable':
                                output_map[col] = 'TRUE'
                            elif col == 'Variant Weight Unit':
                                output_map[col] = 'kg'
                            else:
                                output_map[col] = ""
                        
                        writer.writerow(output_map)
                        decant_idx += 1
        except Exception as e:
            print("Error",e)
            continue

                        
                        
                        
```
