```python
input_df = pd.read_csv(f"{csv_file_path}")
```

```python
input_df = input_df[['Name', 'Lineitem name', 'Lineitem quantity', 'Lineitem price']]
input_df = input_df[input_df['Lineitem name'].str.contains('Decant', 'decant')]
input_df = input_df.groupby(["Name", "Lineitem name", "Lineitem quantity"], as_index=False).agg({"Lineitem price": "max"})
df_expanded = input_df.loc[input_df.index.repeat(input_df['Lineitem quantity'])].reset_index(drop=True)
```

```python
def process_name(default_name):
   
    pattern = r'\s*(For\s+(Women|Men|women|men|man|Man|woman|women).*$)'
    
    print_name = re.sub(pattern, '', default_name, flags=re.IGNORECASE).strip()
    
    return print_name.title()
```

```python
def extract_quantity(text):
    """
    Extract quantity (e.g., 5ml, 10ml) from text.
    Returns normalized format: "NUMBER ML"
    """
    pattern = r'\b(\d+)\s*(ml)\b'
    
    match = re.search(pattern, text, re.IGNORECASE)
    
    if match:
        number = match.group(1)
        unit = match.group(2).upper()
        normalized = f"{number} {unit}"
        return normalized
    
    return ''
```

```python
input_df = df_expanded[['Name', 'Lineitem name']].copy()

input_df['print_name'] = input_df['Lineitem name'].apply(process_name)
input_df['quantity'] = input_df['Lineitem name'].apply(extract_quantity)
#input_df = input_df[~input_df['print_name'].str.contains('Partial Payment', na=False)].reset_index(drop=True)
#input_df = input_df.groupby(["Name", "Lineitem name"], as_index=False).agg({"Lineitem price": "max"})
```
