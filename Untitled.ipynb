```python
import pandas as pd
```

```python
csv_file_path = r"C:\Users\silla\Downloads\orders_export_24_10_v1.csv"
output_file = r"C:\Users\silla\Downloads\print_labels_24_10_v1.csv"
```

```python
input_df = pd.read_csv(f"{csv_file_path}")
```

```python
input_df = input_df[['Name', 'Lineitem name']]
input_df = input_df[input_df['Lineitem name'].str.contains('Decant', 'decant')]
```

```python
def process_name(default_name):
    print_name = default_name.lower().split('for')[0]
    return print_name.title()
```

```python
input_df['print_name'] = input_df['Lineitem name'].apply(process_name)
```
